```python
import string
allAlphabets=list(string.ascii_uppercase)
```

```python
frequency=[]
for each in allAlphabets:
    wordswithEach=list(filter(lambda x:x.startswith(each),names.words()))
    frequency.append(len(wordswithEach))

for every in zip(allAlphabets,frequency):
    print(every) 
```

```python
import matplotlib.pyplot as plt
tic=range(len(allAlphabets))
plt.bar(tic,frequency)
plt.xticks(tic,allAlphabets)
plt.title("Plot of Frequency of Alphabets")
plt.xlabel("Alphabets")
plt.ylabel("Frequency")
plt.savefig("AlphabetsFrequency.png")
plt.show()

```
